```python
#pd.Dataframe directly creates a dataframe, Useful
#creating dataframes with no common columns

df3 = pd.DataFrame({'Name 1': ['Matt','Kyle','Tara','Nora','Brian'],'Final Score': [92,88,84,87,75]})
df4 = pd.DataFrame({'Name 2': ['Tara', 'Kyle', 'Ali'], 'Place of birth': ['Wisconsin','Wisconsin','Illinois']})
```

```python
df3
```

```python
df4
```

```python
pd.merge(df3, df4, left_on='Name 1', right_on='Name 2')
```

```python
#on:Column names to join on. Must be found in both DataFrame objects.
#If not specified and no other join keys given, will use the intersection of the column names in left and right as the join keys
#left_on: Columns in left DataFrame to use as join keys
#right_on: Analogous to left_on for left DataFrame
pd.merge(df3, df4, how='outer',left_on='Name 1', right_on='Name 2')
```

```python
#Tidy Data Definition:

#    Each variable forms a column and contains values
#    Each observation forms a row
#    Each type of observational unit forms a table
    
# Definitions of the above:

#    Variable: A measurement or an attribute. Height, weight, sex, etc.
#    Value: The actual measurement or attribute. 152 cm, 80 kg, female, etc.
#    Observation: All values measure on the same unit. Each person
        
#Principle 1: Column headers are values, not variable names
```
